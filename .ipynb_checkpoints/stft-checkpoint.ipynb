```python
from scipy import signal
from scipy.io import wavfile
import matplotlib.pyplot as plt
import numpy as np
import os
from scipy.signal import remez
import math
from scipy.signal import freqz
```

```python
# read data from wav file
fs, data = wavfile.read('spongebob.wav')
data = data[:,0]

# STFT
amp = 2 * np.sqrt(2)
f, t, Zxx = signal.stft(data, fs, nperseg=10000)
plt.pcolormesh(t, f, np.abs(Zxx), vmin=0, vmax=amp)
plt.title("STFT magnitude")
plt.ylabel("Frequency")
plt.xlabel("Time")
plt.show()
```

```python
f1, f2, f3 = 1000, 4000, 10000
amps = [10, 15, 22]
ls13 = np.linspace(0,int(len(data)/4), len(data)/4)
ls2 = np.linspace(0, int(len(data)/2), len(data)/2)

n1 = np.cos(2*np.pi*f1*ls13)*amps[0]
n2 = np.cos(2*np.pi*f2*ls2)*amps[1]
n3 = np.cos(2*np.pi*f3*ls13)*amps[2]
noises = np.array([])
noises = np.append(noises, n1)
noises = np.append(noises, n2)
noises = np.append(noises, n3)
print(len(noises))
plt.plot(noises)
```

```python
# noised record
n_data = data + noises
f2, t2, Zxx2 = signal.stft(n_data, fs, nperseg=1000)
plt.pcolormesh(t2, f2, np.abs(Zxx2), vmin=0, vmax=amp)
plt.title("STFT magnitude")
plt.ylabel("Frequency")
plt.xlabel("Time")
plt.show()

f, t, Zxx = signal.stft(data, fs, nperseg=1000)
plt.pcolormesh(t, f, np.abs(Zxx), vmin=0, vmax=amp)
plt.title("STFT magnitude")
plt.ylabel("Frequency")
plt.xlabel("Time")
plt.show()
```

```python
fil = signal.remez(fs, [0, 800, 900, 1100, 1200, fs/2], [1, 0, 1], fs = fs)
w1, h1 = freqz(fil)
plt.semilogy((fs*w1)/(2*np.pi),np.abs(h1))
plt.show()

fil2 = signal.remez(int(300*np.pi), [0, 3600, 3800, 4200, 4400, fs/2], [1, 0, 1], fs = fs)
w2, h2 = freqz(fil2)
plt.semilogy((fs*w2)/(2*np.pi),np.abs(h2))
plt.show()

fil3 = signal.remez(int(300*np.pi), [0, 9600, 9800, 10200, 10400, fs/2], [1, 0, 1], fs = fs)
w3, h3 = freqz(fil3)
plt.semilogy((fs*w3)/(2*np.pi),np.abs(h3))
plt.show()
```
